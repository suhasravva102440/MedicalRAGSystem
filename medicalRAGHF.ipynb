```python
# Load the documents

from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import TextLoader

loader = DirectoryLoader('medicaldata/notes', glob="*.txt", show_progress=True, loader_cls=TextLoader)

kb_docs = loader.load()


# Chunk the loaded documents

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)

chunks = text_splitter.split_documents(kb_docs)
```

```python
from langchain_chroma import Chroma

from langchain_huggingface import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-mpnet-base-v2"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
hf = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs,
)
# Initialize a ChromaDB Connection
from langchain_chroma import Chroma

# Initialize the database connection
# If database exist, it will connect with the collection_name and persist_directory
# Otherwise a new collection will be created
db = Chroma(collection_name="vector_database", 
            embedding_function=hf, 
            persist_directory="./chroma_db1_")

from tqdm import tqdm

batch_size = 1000 
# 2. Use a loop to send chunks in segments
for i in tqdm(range(0, len(chunks), batch_size)):
    batch = chunks[i : i + batch_size]
    
    # 3. Add only the current batch
    db.add_documents(documents=batch)
```

```python
# Helper function to join the retrieved chunks

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
# Step 1: Initialize the Chroma DB Connection

from langchain_chroma import Chroma
from langchain_core.runnables import RunnablePassthrough

# Initialize the database connection
# If database exist, it will connect with the collection_name and persist_directory
# Otherwise a new collection will be created
db = Chroma(collection_name="vector_database", 
            embedding_function=hf, 
            persist_directory="./chroma_db1_")

# We can check the already existing values
#print(len(db.get()["ids"]))


# Step 2: Create a Retriever Object 

retriever = db.as_retriever(search_type="similarity", search_kwargs={"k": 3})

# Step 3: Initialize a Chat Prompt Template

from langchain_core.prompts import ChatPromptTemplate

PROMPT_TEMPLATE = """
Answer the question based only on the following context:
{context}
Answer the question based on the above context: {question}.
Provide a detailed answer.
Don’t justify your answers.
Don’t give information not mentioned in the CONTEXT INFORMATION.
Do not say "according to the context" or "mentioned in the context" or similar.
"""

prompt_template = ChatPromptTemplate(
    messages=[
        PROMPT_TEMPLATE
    ]
)

# step 4: Define chatmodel

f = open('keys/google-api.txt')

GOOGLE_API_KEY = f.read()

from langchain_google_genai import ChatGoogleGenerativeAI
# Pass the standard parameters during initialization
google_chat_model = ChatGoogleGenerativeAI(api_key=GOOGLE_API_KEY, 
                                           model="gemini-2.5-flash", 
                                           temperature=1)

from langchain_core.output_parsers import StrOutputParser

parser = StrOutputParser()

# build generator chain now

generator_chain = prompt_template|google_chat_model|parser

rag_chain = { "context" : retriever | format_docs,
              "question" : RunnablePassthrough()} | generator_chain



rag_chain.invoke(" Is there a cure for Alzheimer’s disease?")
```
